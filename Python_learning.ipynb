## Dictionaries - Collection of key-value pairs which is ordered, changeable and indexed.  each key is unique and associated with a value.

```python
my_dict = {"a":"1", "b" : "2", "c" : "3"}
```

```python
# Access the item values based on key
my_dict["a"]
```

```python
# Traverse througn the keys
for x in my_dict:
  print(x) # it will print only keys
```

```python
# Traverse Through values
for x in my_dict.values():
  print(x) # it will print only values
```

```python
# Traverse through both
for x in my_dict.items():
  print(x) # it will print both keys and values
```

```python
my_dict['d'] = '4' # adding new element to dictionary
my_dict
```

```python
# we can even replace values in dict
my_dict['d'] = '5'
my_dict
```

- Nested Dictionary

```python
car1_model = {'Mercedes':1960}
car2_model = {'Audi':1970}
car3_model = {'Ambassador':1980}

car_type = {'car1':car1_model, 'car2':car2_model, 'car3':car3_model}
```

```python
print(car_type)
print(car_type['car1']) # to get 1st value
print(car_type['car1']['Mercedes']) # to get 2nd value
```

# 6. Matplotlib (visualization Library)

## Matplotlib is a plotting library for python and numerical maths extension numpy. it provides object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, wxPython, QL or GTK+.

Some major Pros of matplotlib are --
- Generally easy to get started for simple plots
- support for custom labels and texts
- Great control of every element in a figure
- high-quality output in many formats
- very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# simple example

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# Plotting using matplotlib

plt.scatter(x,y , c = 'g') # c= color , g = green
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title ('2D graph')

# plt.show() // we do not need to use this because we alredy use %matplotlib inline

plt.savefig('Test.png') # to save the giving plot in giving directory
```

```python
z = np.arange(1,100)
a = np.arange(1,10)
z = a*a
plt.plot(a, z, 'ro--')
```

```python
plt.plot(z, a, '+', c = 'g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title ('2D graph')
```

```python
plt.subplot(2,2,3) # no. of rows, no. of coloum, series no.
plt.plot(z, a, '+', c = 'g')
plt.subplot(2,2,1)
plt.plot(z, a, 'b*' , linestyle = 'dashed')
plt.subplot(2,2,2)
plt.plot(z, a, '+', c = 'r')
```

```python
np.pi # value of pi
```

```python
t = np.arange(0, 4*np.pi, 0.1) # 0.1 = stepsize
u = np.sin(t)
plt.plot(t,u)
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.title('sin graph')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.plot(x, y_sin, 'r*')

plt.subplot(2,2,2)
plt.title('cos graph')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.plot(x,y_cos, 'g+')
```

```python
a = np.array([10,20,54,87,10,23,22, 50,52,53,54,54]) # shows density of number, or count of range
plt.hist(a)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation. it usually a scatterplot, a hexbin plot, a 2d histogram or a 2d density plot

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind = 'reg')
```
